```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/content/labrador-retriever-1210559__480.jpg")
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(rgb_img)
plt.show()
```

```python
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_img,cmap='gray')
plt.show()
```

```python
invert_gray_img =255 - gray_img
plt.figure(figsize=(8,8))
plt.imshow(invert_gray_img,cmap='gray')
plt.show()
```

```python
blur_img = cv2.GaussianBlur(invert_gray_img, (21,21), 0)
plt.figure(figsize=(8,8))
plt.imshow(blur_img,cmap='gray')
plt.show()
```

```python
invert_blur_img = 255 - blur_img
plt.figure(figsize=(8,8))
plt.imshow(invert_blur_img,cmap='gray')
plt.show()
```

```python
sketch = cv2.divide(gray_img, invert_blur_img, scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(rgb_img)
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap='gray')
plt.show()
```
