# **Importing Required Libraries**

```python
from google.colab.patches import cv2_imshow

import cv2
```

# **Loading the image**

Here we will read the image that is to be processed.

```python
img = cv2.imread('/content/labrador-retriever-1210559__480.jpg')
cv2_imshow(img)
```

# **Converting image into gray scale image.**
By this we will convert the above image into black & white pixels which is used for creating pencil sketch.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

# **Inverting the image**
'bitwise_not' function is used to make brighter regions lighter and vice-versa so that we can find edges to create a pencil sketch

```python
img_inv = cv2.bitwise_not(img_gray)
cv2_imshow(img_inv)
```

# **Smoothing the image**
It is used so that by increasing the filter size, we can create thin lines for our sketch and also to reduce the noice in the image.

```python
img_smooth = cv2.GaussianBlur(img_inv, (21, 21), sigmaX= 0, sigmaY=0)
cv2_imshow(img_smooth)
```

# **Obtaining the final sketch**
We will be using a blend function 'dodgeV2'

By using this function, it is dividing the the greyscale value of the image by the inverse of blurred image value which highlights the boldest edges.

Note - This technique is used by traditional photographers to print photos from the reels

```python
def dodgeV2(x,y):
  return cv2.divide(x, 255 - y, scale=256)

final_img = dodgeV2(img_gray, img_smooth)
cv2_imshow(final_img)
```
